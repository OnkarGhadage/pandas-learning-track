# Creating Series & DataFrame

## 1. Create a Series from a List

```python
import pandas as pd
```

```python
s = pd.Series([10,20,30,40,50])
s
```

## 2. Create a Series from a Dict

```python
marks = pd.Series({
    "Phy": 88,
    "Chem": 85,
    "Math": 98
})
marks
```

## 3. Create a DataFrame from Lists

```python
df = pd.DataFrame({
    "Name": ["Aaa", "Bbb", "Ccc"],
    "Age": [12, 23, 34],
    "City": ["Yyy", "Xxx", "Zzz"]
})
df
```

## 4. Create a DataFrame from a List of Dicts

```python
data = [
    {"Name": "Aaa", "Age": 22,  "City": "Pune"},
    {"Name": "Ggg", "Age": 32,  "City": "Mumbai"},
    {"Name": "Kkk", "Age": 12,  "City": "Delhi"},
]
df = pd.DataFrame(data)
df
```

## 5. Create a DataFrame from NumPy Arrays

```python
import numpy as np
```

```python
arr = np.array([[10, 20, 30],
               [40, 50, 60]])
df_1 = pd.DataFrame(arr, columns=[1, 2, 3], index=["A", "B"])

df_1
```
