```python
import pandas as pd
```

# Viewing Data  
(`.head()`, `.tail()`, `.sample()`, `.info()`, `.describe()`)

lets setup a sample DataFrame

```python
df = pd.DataFrame({
    "Name": ["Onkar", "Amit", "Sara", "Rohit", "Neha", "Vishal"],
    "Age": [21, 25, 23, 29, 20, 26],
    "City": ["Pune", "Mumbai", "Nashik", "Pune", "Mumbai", "Nagpur"],
    "Salary": [50000, 65000, 55000, 70000, 48000, 62000]
})

df
```

## 1. `.head()` -> First few rows  
Defalt 5 rows but we can randomly give number

```python
df.head()
```

```python
df.head(2) # First 2 rows
```

## 2. `.tail()` -> Last few rows  
Default 5 rows but in this also we can give random number

```python
df.tail()
```

```python
df.tail(2) # Last 2 rows
```

## 3. `.sample()` -> Random rows  
Default only 1 but we can give random number also

```python
df.sample()
```

```python
df.sample(2) # Gives 2 random rows
```

## 4. `.info()` -> Quick summary  
It shows column names, datatypes, notnull values, memory usages, etc  
It is very usefull before cleaning the data

```python
df.info()
```

## 5. `.describe()` -> States summary (numerical columns only)  
It shows mean, std (standard deviation), min/max, etc

```python
df.describe()
```

For all columns we need to pass `include="all"`  
For objects there are some extra contents like unique, top, freq

```python
df.describe(include="all")
```
