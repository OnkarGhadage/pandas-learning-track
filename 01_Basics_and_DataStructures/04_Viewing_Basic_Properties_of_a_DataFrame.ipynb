# Viewing Basic Properties of a DataFrame

## 1. Setup a Sample DataFrame

```python
df = pd.DataFrame({
    "Name": ["Aaa", "Bbb", "Ccc", "Ddd"],
    "Age": [21, 25, 27, 30],
    "City": ["Mumbai", "Pune", "Nashik", "Satara"]
})

df
```

## 2. `.shape` -> Rows & Columns

```python
df.shape
```

(rows, columns)  
(4, 3)   
4 - Rows  
3 - Columns  

## 3. `.dtypes` -> Data type of each columns
  
object - String  
int64 - integer  
float64 - floating point number  

```python
df.dtypes
```

## 4. `.index` -> Row labels

```python
df.index
```

its same line python for loop range() function  
range(4) -> 0, 1, 2, 3.  

## 5. `.columns` -> Column names

```python
df.columns
```

## 5. `.info()`
It showns information of DF.

```python
df.info()
```

## 5. `.describe()`
It shows count, mean, std, min, etc

```python
df.describe()
```
